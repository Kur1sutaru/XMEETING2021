````python
dados['cmap'].value_counts().plot.bar()
````

````python
dados['union'].value_counts().plot.bar()
````

````python
cod_l1000 = dados['L1000'].value_counts().index[0:20]
````

````python
cod_l1000
````

````python
dados.query('L1000 in @cod_l1000')
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(60, 12))
ax = sns.countplot(x = 'L1000', data=dados.query('L1000 in @cod_l1000'), order= dados['L1000'].value_counts().index[0:20], palette='husl')
ax.set_title('L1000')
plt.title("L1000 drug list", fontsize = 16)
plt.xlabel("drug", fontsize = 16)
plt.ylabel("count", fontsize = 16)
plt.show()
````

````python
pd.crosstab(dados['L1000'], dados['cmap'])
````

````python
import seaborn as sns
sns.set(rc={'figure.figsize':(30,12)})
sns.scatterplot(x='L1000', y = 'gene', data=dados)
````
